### <span style="color:black">Exercici 4.</span>

Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast entre si les seves dimensions coincideixen o si una de les matrius té una mida d'1": <span style="color:blue">__[X] Fet__</span>

```python
a = np.repeat((0, 5, 10, 15, 20),3).reshape(5, 3) #Exemple 1
b = np.repeat((0, 1, 2) ,5).reshape(3, 5).T
print(a)
print(b)
print(a + b)
```

```python
a = np.repeat((0, 10, 20, 30), 3).reshape(4, 3) #Exemple 2
b = np.array((0, 1, 2))
print(a)
print(b)
print(a + b)
```

```python
a = np.array((0, 2, 4, 6)).reshape(4,1) #Exemple 3
b = np.array((0, 1, 2))
print(a)
print(b)
print(a * b)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image   
from scipy import ndimage
%matplotlib inline
```

```python
image=mpimg.imread("Tasca 2A_FotoOriginal.jpg") #Leemos la imagen
```

```python
image
```

```python
image.dtype
```

```python
print(image.shape) #vemos su tamaño y la visualizamos
plt.imshow(image)
```

```python
img=np.array(image[25:130,50:175]) #Extraemos el pez amarillo
plt.imshow(img,interpolation='nearest')
print(img.shape)
```

```python
imag=image[200:320] #slicing
plt.imshow(imag)
print(imag.shape)
```

```python
#Eliminamos los canales Verde y Azul

imagen=np.copy(image) # creo una copia de la imagen para preservar la original
imagen[:,:,1]=0 #elimino la información del canal verde igualándolo a cero
imagen[:,:,2]=0 #elimino la información del canal azul igualándolo a cero
plt.title("Imagen Canal Rojo (Verde y Azul eliminados)")
plt.imshow(imagen)
```

```python
mpimg.imsave("Foto_CanalRojo.png",imagen)
```

```python
mpimg.imsave("Foto_PezAmarillo.png",img) 
```
